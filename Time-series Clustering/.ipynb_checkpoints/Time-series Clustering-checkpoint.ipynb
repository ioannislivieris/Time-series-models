```python
# =-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
# Built-in libraries
#
import os
import time
import random
import numpy  as np
import pandas as pd


# =-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
# Visualization libraries
#
import matplotlib.pyplot as plt
import seaborn           as sns

# =-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
# Scipy library
#
from   scipy                   import stats
import scipy.cluster.hierarchy as sch


# =-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
# Tensorflow library
#
import tensorflow                  as tf
from   tensorflow.keras.models     import *
from   tensorflow.keras.layers     import *
from   tensorflow.keras.callbacks  import *
# from   tensorflow.keras.optimizers import *
```

# Import data

```python
# Start timer
#
start = time.time()


# Load dataset
#
data, person_id = [], []
for txt in os.listdir('Data/Walk_MODEL'):
    measure = pd.read_csv('Data/Walk_MODEL/'+txt, sep=' ', header=None).values
    data.append(measure)
    person_id.append(txt.split('-')[-1].split('.')[0])

data = np.asarray(data)
print('[INFO] Dataset loaded')
print('[INFO] Number of examples: ', data.shape[0])



# Stop timer
#
print('[INFO] Time %.2f' % (time.time() - start))
```

## Visualization of series for random individual

```python
# Select random individual
#
idx = 10


# Plot
plt.figure( figsize=(20, 4) )

plt.plot(data[11]);
plt.legend(['X','Y','Z'], frameon=False);

plt.title(person_id[idx]); 
plt.xlabel('Time');
```

## Scaling (Standarization)

```python
### STANDARDIZE DATA ###

for i, s in enumerate(data):
    data[i] = (data[i] - s.mean(axis=0)) / s.std(axis=0)
```

# DIMENSIONALITY REDUCTION

```python
# Utility function for creatign model
#    
def AutoEncoder( data ):
    
    input_layer = Input(shape=(data.shape[1], data.shape[2]))
    
    encoder = TimeDistributed(Dense(200, activation='tanh'))(input_layer)
    encoder = TimeDistributed(Dense(50, activation='tanh'))(encoder)
    #
    latent  = TimeDistributed(Dense(10, activation='tanh'))(encoder)
    #
    decoder = TimeDistributed(Dense(50, activation='tanh'))(latent)
    decoder = TimeDistributed(Dense(200, activation='tanh'))(decoder)

    out = TimeDistributed(Dense(data.shape[2]))(decoder)

    
    autoencoder = Model(inputs=input_layer, outputs=out)
    autoencoder.compile(optimizer='adam', loss='mse')
    
    return autoencoder
```

# Setup AutoEncoder

```python
model = AutoEncoder(data)
model.summary()
```

## Training

```python
# Earlystopping
#
earlystopping = EarlyStopping(patience             = 20, 
                              verbose              = 1, 
                              min_delta            = 0.001, 
                              monitor              = 'val_loss', 
                              mode                 = 'auto', 
                              restore_best_weights = True)


# Learning rate adjustment
#
lrs_scheduler  = ReduceLROnPlateau(monitor     = 'val_loss', 
                                    factor      = 0.5,
                                    patience    = 10)
```

```python
# Start timer
#
start = time.time()


# Train AutoEncoder
#
model.fit(data[:15], data[:15], 
          epochs    = 200, 
          verbose   = 1, 
          callbacks = [earlystopping, lrs_scheduler],
          validation_data = (data[15:],data[15:]), 
          )


# Stop timer
#
print('[INFO] Time %.2f' % (time.time() - start))
```

## Evaluation

```python
# Select random individual
#
idx = 17

# Get prediction for random individual
#
pred = model.predict( data[ [idx] ] )

# Calculate MSE
#
MSE = np.mean(np.power(data[ [idx] ] - pred, 2), axis=2)


# Plot
#
plt.figure(figsize=(20, 3))

plt.scatter(range(data.shape[1]), MSE)
plt.title('Reconstruction error - Person ID: ' + person_id[idx])
plt.xlabel('Time'); 
plt.ylabel('MSE');
```

## Create DataFrame with Reconstruction errors

```python
df = np.mean(np.power(data[15:] - model.predict(data[15:]), 2), axis=2)
```

# Pearson Correlation CLUSTERING

```python
# Compute correlation matrix
#
corr = np.corrcoef(df)



plt.figure(figsize = (8, 6))
sns.set(font_scale = 1)

sns.heatmap(data        = corr,
            cbar        = True,
            annot       = True,
            square      = True,
            fmt         = '.2f',
            annot_kws   = {'size': 14},
            #
            yticklabels = person_id[15:],
            xticklabels = person_id[15:])


plt.title('Pearson correlation coefficients', size = 14, weight='bold')
plt.tight_layout()
plt.show()
```

```python
# Hierachical Clustering
#
d = sch.distance.pdist(corr)
L = sch.linkage(d, method='ward')

ind        = sch.fcluster(L, d.max(), 'distance')
dendrogram = sch.dendrogram(L, no_plot=True)

df = [df[i] for i in dendrogram['leaves']]
labels = [person_id[15:][i] for i in dendrogram['leaves']]
corr = np.corrcoef(df)




plt.figure(figsize = (8, 6))
sns.set(font_scale = 1)

sns.heatmap(data        = corr,
            cbar        = True,
            annot       = True,
            square      = True,
            fmt         = '.2f',
            annot_kws   = {'size': 14},
            #
            yticklabels = person_id[15:],
            xticklabels = person_id[15:])


plt.title('Hierachical Clustering', size = 14, weight='bold')
plt.tight_layout()
plt.show()
```

## Plot dentrogram

```python
plt.figure(figsize=(8,6));

# Plot dendrogram
#
dendrogram = sch.dendrogram(L, labels=person_id[15:])

# Plot threshold
#
plt.axhline(d.max(), color='black');
```

# Kolmogorov-Smirnov CLUSTERING

```python
def KScoeff(df):
    
    ks_matrix = np.zeros((len(df),len(df)))
    for i in range(len(df)):
        for j in range(i+1,len(df)):
            ks_test = stats.ks_2samp(df[i], df[j])
            ks_matrix[i,j] = ks_test.statistic
            ks_matrix[j,i] = ks_test.statistic
            
    return ks_matrix
```

```python
# Compute Kolmogorov-Smirnov matrix
#
ks_matrix = KScoeff(df)





plt.figure(figsize = (8, 6))
sns.set(font_scale = 1)

sns.heatmap(data        = ks_matrix,
            cbar        = True,
            annot       = True,
            square      = True,
            fmt         = '.2f',
            annot_kws   = {'size': 14},
            #
            yticklabels = person_id[15:],
            xticklabels = person_id[15:])


plt.title('Kolmogorov-Smirnov Clustering', size = 14, weight='bold')
plt.tight_layout()
plt.show()
```

## Hierachical Clustering

```python
# Hierachical Clustering
#
d = sch.distance.pdist(ks_matrix)
L = sch.linkage(d, method='ward')

ind        = sch.fcluster(L, d.max(), 'distance')
dendrogram = sch.dendrogram(L, no_plot=True)

df = [df[i] for i in dendrogram['leaves']]
labels = [person_id[15:][i] for i in dendrogram['leaves']]
ks_matrix = KScoeff(df)




plt.figure(figsize = (8, 6))
sns.set(font_scale = 1)

sns.heatmap(data        = ks_matrix,
            cbar        = True,
            annot       = True,
            square      = True,
            fmt         = '.2f',
            annot_kws   = {'size': 14},
            #
            yticklabels = person_id[15:],
            xticklabels = person_id[15:])


plt.title('Hierachical Clustering', size = 14, weight='bold')
plt.tight_layout()
plt.show()
```

## Plot dendrogram

```python
plt.figure(figsize=(8,6));

# Plot dendrogram
#
dendrogram = sch.dendrogram(L, labels=person_id[15:])

# Plot threshold
#
plt.axhline(d.max(), color='black');
```
