# Parameters

```python
Smoothing = True


Lag     = 20
Horizon = 5
```

# Import data

```python
# Start timer
#
start = time.time()


# Load dataset
#
df = pd.read_csv('Data/PV_Elec_Gas2.csv')
print('[INFO] Dataset loaded')


# Stop timer
#
print('[INFO] Time %.2f' % (time.time() - start))


df.head()
```

## Preprocess

```python
df.sort_values('Date', inplace=True)
df.set_index('Date', inplace=True)


# Calculate power
#
df['Power'] = df.cum_power.diff()

# Remove unuseful columns
#
df.drop(['cum_power'], axis=1, inplace=True)

# Remove empty instances
#
df.dropna(inplace=True)
```

## Visualization

```python
plt.figure(figsize=(20, 4))
#
df.Power.plot(color='red')
df.Elec_kW.plot(color='orange')
df.Gas_mxm.plot(color='green')

plt.legend(frameon=False)
```

## Use kalman filter to smooth data (only for visualization)

```python
# Setup Kalman-Smoother
#
smoother = KalmanSmoother(component       = 'level_longseason', 
                          component_noise = {'level':0.1, 'longseason':0.1}, 
                          n_longseasons   = 365)

smoother.smooth(df[['Elec_kW','Gas_mxm','Power']].T);


# Create DataFrame with smoothed data
#
Smoothed_df = pd.DataFrame( [] )

for i, feature in enumerate(df.columns):
    Smoothed_df[ feature ] = smoother.smooth_data[i]
    
    
plt.figure(figsize=(20, 4));
#
Smoothed_df.Power.plot(color='red');
Smoothed_df.Elec_kW.plot(color='orange');
Smoothed_df.Gas_mxm.plot(color='green');

plt.legend(frameon=False);
```

## Compare traditional vs smoothed data

```python
for feature in df.columns:
    plt.figure( figsize=(20, 3) )
    df[ feature ].plot(color='tab:blue')
    Smoothed_df[ feature ].plot(color='tab:orange', linestyle='--')
    
    
    plt.title('Feature: {}'.format(feature))
    plt.legend(['Traditional', 'Smoothed'], frameon=False)
```

# Training/Testing data

## Split data

```python
trainX, testX, trainY, testY = train_test_split(df[['Elec_kW','Gas_mxm','Power']].values, df[['Power']].values, 
                                                test_size = 0.2, 
                                                shuffle   = False)
```

## Create 'Input' data

```python
### SLICE TRAIN/VAL DATA INTO EQUAL SLIDING WINDOWS ###


trainX = create_windows(data         = trainX, 
                        window_shape = Lag, 
                        end_id       = -Horizon)


testX = create_windows(data         = testX, 
                        window_shape = Lag, 
                        end_id       = -Horizon)
```

## Create 'Output' data

### Use Kalman filter to smooth data - Only apply on target-series

```python
# Setup Kalman-Smoother
#
smoother = KalmanSmoother(component       = 'level_longseason', 
                          component_noise = {'level':0.1, 'longseason':0.1}, 
                          n_longseasons   = 365);

smoother.smooth( trainY.T );
```

```python
# Create sliding window on smooth training target-series
#
trainY = create_windows(data  = smoother.smooth_data.T, 
                                window_shape = Horizon, 
                                start_id     = Lag)


# Create sliding window on testing target-series
#
testY = create_windows(data         = testY, 
                       window_shape = Horizon, 
                       start_id     = Lag)
```

## Scaling

```python
# Setup scaler
#
scalerX = StandardScaler()

trainX = scalerX.fit_transform(trainX.reshape(-1, trainX.shape[-1])).reshape(trainX.shape)
testX  = scalerX.transform(testX.reshape(-1, testX.shape[-1])).reshape(testX.shape)
```

```python
# Setup scaler
#
scalerY = StandardScaler()

trainY = scalerY.fit_transform(trainY.reshape(-1, trainY.shape[-1])).reshape(trainY.shape)
```

# Forecasting Model

```python
def CreateModel( param ):
   
    model = Sequential()
    model.add(LSTM(param['unit'], activation=param['act']))
    model.add(RepeatVector(param['Horizon']))
    model.add(LSTM(param['unit'], activation=param['act'], return_sequences=True))
    model.add(TimeDistributed(Dense(1)))
    
    opt_choices = {'adam': Adam(),
                   'rms':  RMSprop()}
    
    opt = opt_choices[ param['opt'] ]
    opt.lr = param['lr'] 
    
    model.compile(opt, 'mse')
    
    return model
```

## Model parameters

```python
### CREATE GRID FOR HYPERPARAM SEARCH ###

param_grid = {
    'unit':       [64, 128], 
    'lr':         [1e-2, 1e-3], 
    'act':        ['elu', 'relu', 'tanh'], 
    'opt':        ['adam'],
    'epochs':     200,
    'batch_size': 512,
    'Horizon':    Horizon
}
```

## Training & Hyperparameter optimization

```python
# Earlystopping
#
earlystopping = EarlyStopping(patience             = 20, 
                              verbose              = 0, 
                              min_delta            = 0.001, 
                              monitor              = 'val_loss', 
                              mode                 = 'auto', 
                              restore_best_weights = True)                          



# Learning rate adjustment
lrs_scheduler  = ReduceLROnPlateau(monitor     = 'val_loss', 
                                    factor      = 0.75,
                                    patience    = 10)
```

```python
hypermodel = CreateModel

kgs = KerasGridSearch(hypermodel        = hypermodel, 
                      param_grid        = param_grid, 
                      monitor           = 'val_loss', 
                      greater_is_better = False, 
                      tuner_verbose     = 1)

kgs.search(trainX, trainY, 
           validation_split = 0.2,
           callbacks       = [earlystopping, lrs_scheduler])
```

```python
print('[INFO] Best score: ', kgs.best_score)



print('[INFO] Optimized hyperparameters\n')
for (parameter, value) in kgs.best_params.items():
    if ( isinstance(value, float) ):
        print(' >%25s: %.3f' % (parameter,value))
    else:
        print(' >%25s: %s' % (parameter,value))
```

# Evaluation

```python
# Get predictions
#
pred = kgs.best_model.predict( testX )
pred = scalerY.inverse_transform(pred.reshape(-1, pred.shape[-1])).reshape(pred.shape)
```

## Evaluate the performance of the developed model against Naive model

```python
Performance_Foresting_Model = {'Horizon': [], 'RMSE': [], 'MAE': [], 'SMAPE': [], 'R2' : []}
Performance_Naive_Model     = {'Horizon': [], 'RMSE': [], 'MAE': [], 'SMAPE': [], 'R2' : []}


for i in range( Horizon ):
        
    # Store results (Performance metrics)
    #
    Performance_Foresting_Model['Horizon'] += [i+1]
    Performance_Foresting_Model['RMSE']    += [ rmse(testY[:,i,0], pred[:,i,0]) ]
    Performance_Foresting_Model['MAE']     += [ mae(testY[:,i,0], pred[:,i,0]) ]
    Performance_Foresting_Model['SMAPE']   += [ smape(testY[:,i,0], pred[:,i,0]) ]
    Performance_Foresting_Model['R2']      += [ R2(testY[:,i,0], pred[:,i,0]) ]
    
    # Store results (Performance metrics)
    #
    Performance_Naive_Model['Horizon'] += [i+1]
    Performance_Naive_Model['RMSE']    += [ rmse(testY[1:,i,0], testY[:-1,0,0]) ]
    Performance_Naive_Model['MAE']     += [ mae(testY[1:,i,0], testY[:-1,0,0]) ]
    Performance_Naive_Model['SMAPE']   += [ smape(testY[1:,i,0], testY[:-1,0,0]) ]
    Performance_Naive_Model['R2']      += [ R2(testY[1:,i,0], testY[:-1,0,0]) ]
    
    
Performance_Foresting_Model = pd.DataFrame( Performance_Foresting_Model )
Performance_Foresting_Model.set_index('Horizon', inplace=True)

Performance_Naive_Model = pd.DataFrame( Performance_Naive_Model )
Performance_Naive_Model.set_index('Horizon', inplace=True)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize=(12, 10))

Performance_Foresting_Model['RMSE'].plot(ax=ax[0][0])
Performance_Foresting_Model['MAE'].plot(ax=ax[0][1])
Performance_Foresting_Model['SMAPE'].plot(ax=ax[1][0])
Performance_Foresting_Model['R2'].plot(ax=ax[1][1])

Performance_Naive_Model['RMSE'].plot(ax=ax[0][0], color='tab:orange')
Performance_Naive_Model['MAE'].plot(ax=ax[0][1], color='tab:orange')
Performance_Naive_Model['SMAPE'].plot(ax=ax[1][0], color='tab:orange')
Performance_Naive_Model['R2'].plot(ax=ax[1][1], color='tab:orange')

ax[0][0].set_ylabel('RMSE');
ax[0][1].set_ylabel('MAE');
ax[1][0].set_ylabel('SMAPE');
ax[1][1].set_ylabel('R2');


ax[0][0].set_title('RMSE');
ax[0][1].set_title('MAE');
ax[1][0].set_title('SMAPE');
ax[1][1].set_title('R2');

fig.legend(['Naive', 'Forecasting model'], fontsize=14, frameon=False);
```
